```python
import wiki_plot_extractor
```

```python
for s in range(1, 9):
    summaries = wiki_plot_extractor.get_episode_summaries(f'../wiki_html/Game of Thrones season {s}.html')
    with open(f'../wiki_summaries/GoT_s{s}.txt', 'w') as f:
        for ep in range(len(summaries)):
            f.write(summaries[ep])
            if ep < len(summaries) - 1:
                f.write('\n')
```
